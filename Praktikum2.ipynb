```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Ensure the file name matches the uploaded file and is in the correct directory.
# If the file was uploaded directly, it might be in the current directory.
dpath = 'Titanic-Dataset-fixed.csv'
df = pd.read_csv(dpath)
df.head()

df = df[['Survived', 'Pclass', 'Age', 'Sex', 'Cabin']]
df.head()
```

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df['Sex'] = le.fit_transform(df['Sex']) # proses encoding
df['Cabin'] = le.fit_transform(df['Cabin']) # proses encoding
```

```python
std = StandardScaler()
df['Age'] = std.fit_transform(df[['Age']])
```
